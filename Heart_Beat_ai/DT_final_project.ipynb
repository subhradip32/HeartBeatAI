# creating model

```python
from pycaret.classification import *
s = setup(data = df,target = "column12")
```

```python
cm = compare_models()
```

```python
model = create_model("lr",verbose=False)
```

```python
newPredictions = predict_model(model, data = pd.DataFrame.from_dict({"column0":[19.0],"column1":[1.0],"column2":[57.00],"column3":	[0],	"column4":[0.160],
                                                                    "column5":	[22.000],"column6":	[4.310],"column7":[12.00],"column8":[1.000]
                                                                    ,"column9":[0.857],"column11":[1]	}))
newPredictions
```
